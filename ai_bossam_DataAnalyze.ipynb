```python
from matplotlib import pyplot as plt
import pandas as pd
import csv
import numpy as np
```

```python
train_slot_loss=[]
train_intent_loss=[]
train_elmo_f=open('C:/Users/user/ai_bossam_elmo/output_elmo_train.csv', 'r', encoding='utf-8')
rdr=csv.reader(train_elmo_f)
for line in rdr:
    try:
        slot_loss=round(float(line[0]), 3)
        intent_loss=round(float(line[1]), 3)
        train_slot_loss.append(slot_loss)
        train_intent_loss.append(intent_loss)
    except:
        continue
train_elmo_f.close()
```

```python
val_slot_loss=[]
val_intent_loss=[]
val_elmo_f=open('C:/Users/user/ai_bossam_elmo/output_elmo_val.csv', 'r', encoding='utf-8')
rdr=csv.reader(val_elmo_f)
for line in rdr:
    try:
        slot_loss=round(float(line[0]), 3)
        intent_loss=round(float(line[1]), 3)
        val_slot_loss.append(slot_loss)
        val_intent_loss.append(intent_loss)
    except:
        continue
val_elmo_f.close()
```

```python
eval_slot_loss=[]
eval_intent_loss=[]
eval_elmo_f=open('C:/Users/user/ai_bossam_elmo/output_elmo_eval.csv', 'r', encoding='utf-8')
rdr=csv.reader(eval_elmo_f)
for line in rdr:
    try:
        slot_loss=round(float(line[0]), 3)
        intent_loss=round(float(line[1]), 3)
        eval_slot_loss.append(slot_loss)
        eval_intent_loss.append(intent_loss)
    except:
        continue
eval_elmo_f.close()
```

```python
print(len(train_slot_loss))
print(len(train_intent_loss))
```

```python
plt.plot(np.arange(30), train_slot_loss, color='red')
plt.plot(np.arange(30), val_slot_loss, color='green')
plt.plot(np.arange(30), eval_slot_loss, color='blue')
plt.title('elmo slot loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

```python
plt.plot(np.arange(30), train_intent_loss, color='red')
plt.plot(np.arange(30), val_intent_loss, color='green')
plt.plot(np.arange(30), eval_intent_loss, color='blue')
plt.title('elmo intent loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.show()
```

```python
elmo_val_slot_loss=val_slot_loss
elmo_val_intent_loss=val_intent_loss
elmo_eval_slot_loss=eval_slot_loss
elmo_eval_intent_loss=eval_intent_loss
```
